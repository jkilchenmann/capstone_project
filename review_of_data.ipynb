## Step 4:  Determine Unique Values and Visualize Them

```python
# Show number of unique values for each attribute
df.nunique()
```

```python
# Show departure airports alphabetically by code
from_codes = df['From IATA Code'].unique()
sorted_from_codes = sorted(from_codes)
print(f"From Airports: {sorted_from_codes}")
```

```python
# Show arrival airports alphabetically by code
df['To IATA Code'].unique()
to_codes = df['To IATA Code'].unique()
sorted_to_codes = sorted(to_codes)
print(f"To Airports: {sorted_to_codes}")
```

```python
# Count how many times each airport was departed for the top 10 airports with the most flights
from_counts = df['From IATA Code'].value_counts().head(10)
print(from_counts)
```

```python
# Create a bar chart showing the number of flights from the top 10 departure airports
plt.figure(figsize=(12, 6))
from_counts.plot(kind='bar', color='coral')
plt.title('Top 10 Departure Airports by Number of Flights')
plt.xlabel('From IATA Code')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# For the top 10 departure airports, count the number of unique destinations
unique_to_counts = df[df['From IATA Code'].isin(from_counts.index)].groupby('From IATA Code')['To IATA Code'].nunique().sort_values(ascending=False)
print(unique_to_counts)
```

```python
# Create a bar chart showing the number of unique destinations for the top 10 departure airports
plt.figure(figsize=(12, 6))
unique_to_counts.plot(kind='bar', color='coral')
plt.title('Top 10 Departure Airports by Number of Unique Destinations')
plt.xlabel('From IATA Code')
plt.ylabel('Number of Unique Destinations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Get top 10 "From IATA Code" airports by number of flights
top_airports = df['From IATA Code'].value_counts().head(10).index

# Bucket distances into 100-mile bins
df['Distance Bucket'] = (df['Distance'] // 100) * 100

# Create histograms in a grid
fig, axes = plt.subplots(5, 2, figsize=(14, 18))  # 5 rows, 2 columns = 10 plots
axes = axes.flatten()  # Flatten the 2D array of axes for easy iteration

for i, airport in enumerate(top_airports):
    ax = axes[i]
    subset = df[df['From IATA Code'] == airport]
    subset['Distance Bucket'].value_counts().sort_index().plot(
        kind='bar', ax=ax, color='skyblue'
    )
    ax.set_title(f'Flight Distances from {airport}')
    ax.set_xlabel('Distance Bucket (miles)')
    ax.set_ylabel('Number of Flights')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Create 100-mile distance buckets
df['Distance Bucket'] = (df['Distance'] // 100) * 100

# Set up a 5x2 grid for subplots
fig, axes = plt.subplots(5, 2, figsize=(14, 18))
axes = axes.flatten()

# Create one histogram for each of the top 10 airports
for i, airport in enumerate(top_airports):
    ax = axes[i]
    subset = df[df['From IATA Code'] == airport]
    
    # Group by distance bucket and count unique destinations
    bucket_counts = (
        subset.groupby('Distance Bucket')['To IATA Code']
        .nunique()
        .sort_index()
    )
    
    # Plot bar chart
    bucket_counts.plot(kind='bar', ax=ax, color='mediumseagreen')
    ax.set_title(f'Unique Destinations from {airport}')
    ax.set_xlabel('Distance Bucket (miles)')
    ax.set_ylabel('Unique Destinations')
    ax.tick_params(axis='x', rotation=45)

# Clean up layout
plt.tight_layout()
plt.show()
```
